### Train PCA and Discuss Number of Components

```python
## Your Answer Here
pca = PCA()
pca.fit(nhanes_scaled)

# Determine the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Decide on the number of components
cumulative_variance = np.cumsum(explained_variance_ratio)
num_components = np.argmax(cumulative_variance >= 0.95) + 1  # 95% variance threshold
print(f"Number of components chosen: {num_components} (explains ~95% of variance)")
```

### Barplot of Components

```python
## Your Answer Here
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, alpha=0.7, label='Individual explained variance')
plt.step(range(1, len(cumulative_variance) + 1), cumulative_variance, where='mid', label='Cumulative explained variance')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Principal Components')
plt.legend(loc='best')
plt.show()
```

### How many components will you use to fit a supervised learning model?

```python
## Your Answer Here
# Decision to use "num_components" based on cumulative variance
print(f"Using {num_components} components for supervised learning because it retains ~95% of the dataset's variance.")

# Transform the data with the chosen number of components
pca_chosen = PCA(n_components=num_components)
nhanes_pca = pca_chosen.fit_transform(nhanes_scaled)
```

I choose the K-Means algorithm for clustering due to its simplicity and efficiency in handling large datasets like NHANES. K-Means minimizes intra-cluster variance, making it effective for identifying distinct groups.

### Cluster nhanes

```python
## Your Answer Here
# Determine the optimal number of clusters using the elbow method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(nhanes_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.grid()
plt.show()

# Based on the elbow plot, choose the number of clusters (e.g., 3)
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(nhanes_scaled)
labels = kmeans.labels_

print(f"Chosen number of clusters: {num_clusters}. The K-Means algorithm identifies {num_clusters} groups based on feature similarities.")
```
